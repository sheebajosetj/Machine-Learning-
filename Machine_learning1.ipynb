```python
import pandas as pd
dataset = pd.DataFrame(data = {"area_sq_ft": [1000,750,250,300], 'Price':[100000,75000,35000,30000]})
display(dataset)
```

```python
x = dataset['area_sq_ft'].values.reshape(-1,1) #The number of data points on which we want to train the model,followed by the no of features we want to train on
y = dataset['Price'].values.reshape(-1,1)
display(x)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[2]])
```
